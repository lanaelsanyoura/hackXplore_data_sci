```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
# Enable inline plotting
%matplotlib inline
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/lanaelsanyoura/hackXplore_data_sci/main/conposcovidloc.csv')
df
```

```python
# Update so we're only using the needed columns
# TODO
needed_columns = ['Case_Reported_Date', 'Age_Group', 'Client_Gender', 'Outcome1', 'Reporting_PHU_City', 'Reporting_PHU_Latitude',	'Reporting_PHU_Longitude' ]
df = df.loc[:, needed_columns]
df
```

```python
# Introduce the Pandas pd.to_datetime object
date = pd.to_datetime('2020-12-01')
date >=  pd.to_datetime('2020-12-12')
```

```python
# Turn the Case_Reported_Date into a pandas.Datetime object
# TODO
df['Case_Reported_Date'] = pd.to_datetime(df['Case_Reported_Date'])
df
```

```python
# Compute the total number of cases per Date
# TODO
# HINT: use df.groupby(), and count
total_cases = df.groupby('Case_Reported_Date').agg({'Age_Group': 'count'})
total_cases.columns = ['total_cases']
total_cases = total_cases.reset_index()
total_cases
```

```python
# Plot the total cases over time 
# TODO
fig, ax = plt.subplots(figsize=(15,6))
ax = sns.lineplot(data=total_cases, x='Case_Reported_Date', y = 'total_cases', ax=ax)
ax.set_title('Positive Cases Per Day 2020')
```

```python
# Compute the total number of cases per Date AND Age_Group
#TODO
# HINT: use df.groupby(), and count
cases = df.groupby(['Case_Reported_Date', 'Age_Group' ]).agg({'Outcome1': 'count'})
cases.columns = ['total_cases']
cases = cases.reset_index()
cases
```

```python
# Plot the total cases over time 
fig, ax = plt.subplots(figsize=(15,6))
ax = sns.lineplot(data=cases, x='Case_Reported_Date',hue='Age_Group',  y = 'total_cases', ax=ax)
ax.set_title('Positive Cases Per Day 2020')
```

```python
# Compute the total number of cases per Date
# HINT: use df.groupby(), and count
city_cases = df.groupby(['Case_Reported_Date', 'Age_Group', 'Reporting_PHU_City' ]).agg({'Outcome1': 'count'})
city_cases.columns = ['total_cases']
city_cases = city_cases.reset_index()
city_cases
```

```python
# Plot the total cases over time 
ax = sns.relplot(kind='line', col='Reporting_PHU_City', data=city_cases[city_cases['Reporting_PHU_City'].isin(['Toronto', 'Mississauga','London'])], x='Case_Reported_Date',hue='Age_Group',  y = 'total_cases')
plt.title('Positive Cases Per Day 2020')
```
